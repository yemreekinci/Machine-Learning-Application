```python
import pandas as pd

# Veriyi yükle
df = pd.read_parquet("dataOutlier.parquet")

# Price sütununu sayısal türe dönüştür
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Eksik değerleri düşür
df = df.dropna(subset=['Price'])

# Encoding işlemleri
# Brand için Target Encoding
encoding_map = df.groupby('Brand')['Price'].mean()
df['Brand_encoded'] = df['Brand'].map(encoding_map)

# Model için Target Encoding
encoding_map = df.groupby('Model')['Price'].mean()
df['Model_encoded'] = df['Model'].map(encoding_map)

# Transmission için One-Hot Encoding
df = pd.get_dummies(df, columns=['Transmission'], drop_first=True)

# Fuel için Target Encoding
encoding_map = df.groupby('Fuel')['Price'].mean()
df['Fuel_encoded'] = df['Fuel'].map(encoding_map)

# Veriyi kaydet
df.to_parquet('dataEncoding.parquet', index=False)
```
