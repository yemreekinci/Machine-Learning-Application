```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_parquet("../PreprocessedData/dataEncoding.parquet")
skewness = df['Price'].skew()
print("Çarpıklık:", skewness)

sns.histplot(df['Price'], kde=True)
plt.show()
```
