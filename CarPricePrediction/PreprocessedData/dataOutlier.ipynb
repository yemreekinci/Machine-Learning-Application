```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_parquet("dataCleaned.parquet")

# Z-score
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))  # Sayısal sütunlar
abs_z_scores = np.abs(z_scores)

# Z-score
df = df[(abs_z_scores < 3).all(axis=1)]

display(df)
display(df.dtypes)

df.to_parquet('dataOutlier.parquet', index=False)
```
